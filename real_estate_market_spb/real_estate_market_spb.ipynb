### 2.1. Приведем столбцы датафрема к новым типам и оптимизируем значения в столбце last_price

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y.%m.%d')  #изменение формата даты
data['is_apartment'] = data['is_apartment'].astype('bool') #изменение типа данных на булев
data['last_price'] = round(data['last_price']/1000, 2) #переведем стоимость в тыс. руб.
```

### 2.2. Проверим датафрейм на пропуски, дубликаты и аномалии

```python
print(data.isna().sum())
```

В столбце balcony есть пропуски (NaN), скорее всего эти пропуски появились в связи с отсутствием данных или неверной их обработкой. Заменяем пропуски на 0 (т.е. отсутсвие балкона) и меняем тип данных на int (количество балконов может быть только целым числом). В столбце locality_name пропуски в названиях населенных пунктов заменим на 'неизвестно'.

```python
data['balcony'] = data['balcony'].fillna(0).astype('int') #заменим пропуски на 0,меням типданных на int
data['locality_name'] = data['locality_name'].fillna('неизвестно') #заменим их на "неизвестно"
```

Данные в столбцах airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest заменить не предстовляется возможным. Поэтому оставлем их в первоначальном виде. Также нет возможности заменить пропуски в столбце days_exposition, оставляем без изменений.


Для заполнения пропусков в столбцах living_area и kitchen_area вычислим коэффициенты отношения площади кухни и жилой площади к общей площади квартиры. Далее заполним соответствующе столбцы произведением общей площади на эти коэффициенты.  


```python
#находим коэффициент отношения жилой площади к общей площади
k_living = data.loc[(~data['living_area'].isna()), 'total_area'].median()/\
data.loc[(~data['living_area'].isna()), 'living_area'].median()

#заменяем пропуски произведением общей площади на полученный коэффициент
data.loc[(data['living_area'].isna()), 'living_area'] = (data['total_area'] / k_living)
print(data[data['living_area'].isna()].head())
```

```python
#находим коэффициент отношения площади кухни к общей площади
k_kitchen = data.loc[(~data['kitchen_area'].isna()), 'total_area'].median()/\
data.loc[(~data['kitchen_area'].isna()), 'kitchen_area'].median()

#заменяем пропуски произведением общей площади на полученный коэффициент
data.loc[(data['kitchen_area'].isna()), 'kitchen_area'] = (data['total_area'] / k_kitchen)
print(data[data['kitchen_area'].isna()].head())
```

В столбце с количеством этажей количество строк с пропуками незначительно, поэтому можем их удалить.  


```python
data = data.dropna(subset=['floors_total'])
data['floors_total'] = data['floors_total'].astype('int') #меняем тип данных на int
data = data.reset_index(drop=True) #выравниваем индексы
```

```python
print(data['floors_total'].value_counts()) #находим уникальные значения

data = data[(data['floors_total']< 28)] #удаляем те строки, в которых количество этажей больше 28
data = data.reset_index(drop=True) #восстанавливаем индексацию
```

Удалим только строки, в которых указаны дома более 28 этажей, чтобы избежать искажения данных.

При рассмотрении данных в столбце ceiling_height были замечены аномально большие значения. Предполагаем, что у этих данных другая мера исчисления (не метры). Чтобы исправить ситуацию, умножим аномальные значения на 0.1.
Создадим функцию ceiling_height_change чтобы изменить слишком большие значения высоты потоков

```python
#создадим функцию, которая выберет в столбце ceiling_height значения более 10 и умнодит их на 0.1

def ceiling_height_change(height):
    if height >= 10:
        return height * 0.1
    return height

data['ceiling_height'] = data['ceiling_height'].apply(ceiling_height_change)
```

Предположим, что высота потолков может зависеть от этажности дома. Поэтому заполним пропуски средним по этажности.

```python
for f in data['floors_total'].unique():
    data.loc[(data['floors_total'] == f) &
             (data['ceiling_height'].isna()),'ceiling_height']=\
    data.loc[(data['ceiling_height'] == f),'ceiling_height'].mean()

data['ceiling_height'] = round(data['ceiling_height'], 1) #округлим значения высоты потолков

print(data['ceiling_height'].value_counts())
```

Согласно стандартам минмальная высота потолка в России 2.5м. Таких значений в датафреме менее 10%, поэтому удаляем их. Также маловероятно, что потолки в домах будут более 7 метров, таких значений также менее 10%, поэтому удалям их.
